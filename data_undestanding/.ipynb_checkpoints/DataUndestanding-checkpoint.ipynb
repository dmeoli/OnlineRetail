## Data Cleaning and Integration

### Products

First of all, it seems that there are some null values in the *ProdDescr* column so, in order to clean in this sense, we need to remove records for which the *ProdID* appear just one time with null description:

```python
prods = df.groupby('ProdID')['ProdDescr'].describe().reset_index()
prods
```
